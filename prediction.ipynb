```python
# %% [markdown]
# ## 1. 라이브러리 임포트 및 데이터 불러오기

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# %%
# CSV 파일 불러오기
df = pd.read_csv("Dataset_Combined.csv")

# 컬럼명을 소문자로 통일 (예: state, declarationtype, incidenttype 등)
df.columns = df.columns.str.lower()

# 데이터의 앞부분과 정보 확인
print("데이터 샘플:")
display(df.head())

print("\n데이터 정보:")
df.info()

# %%
# state, declarationtype, incidenttype의 고유값 보기
print("Unique state values:", df['state'].unique())
print("Unique declarationtype values:", df['declarationtype'].unique())
print("Unique incidenttype values:", df['incidenttype'].unique())

# %% [markdown]
# ## 2. 모델 학습을 위한 데이터 전처리
#
# 여기서는 사용자 입력과 예측에 사용할 피처로 아래 5개를 선택합니다.
# - **state** (문자형 → Label Encoding)
# - **month** (숫자형)
# - **maxtemp** (숫자형)
# - **mintemp** (숫자형)
# - **precipitation** (숫자형)
#
# 타겟은 **incidenttype** (Label Encoding)

# %%
# 사용할 feature 및 target 컬럼 선택
features = ['state', 'month', 'maxtemp', 'mintemp', 'precipitation']
target = 'incidenttype'

# 일부 행에 결측치나 이상치가 있을 수 있으므로 우선 결측치 제거 (필요시 추가 전처리 진행)
df = df.dropna(subset=features + [target])

# 입력 데이터(X)와 타겟(y) 분리
X = df[features].copy()
y = df[target].copy()

# %%
# 범주형 변수 'state'를 LabelEncoder로 인코딩
le_state = LabelEncoder()
X['state'] = le_state.fit_transform(X['state'])

# 타겟인 incidenttype도 인코딩 (나중에 복원하기 위함)
le_incident = LabelEncoder()
y_encoded = le_incident.fit_transform(y)
num_classes = len(le_incident.classes_)
print("예측할 incidenttype 클래스:", le_incident.classes_)

# %%
# 숫자형 피처에 대해 StandardScaler 적용 (month, maxtemp, mintemp, precipitation)
scaler = StandardScaler()
X[['month', 'maxtemp', 'mintemp', 'precipitation']] = scaler.fit_transform(X[['month', 'maxtemp', 'mintemp', 'precipitation']])

# %%
# 학습/검증 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# %% [markdown]
# ## 3. 딥러닝 분류 모델 생성 및 학습

# %%
# 간단한 분류 모델 구성
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(32, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

model.summary()

# %%
# 모델 학습 (에포크 수, 배치 사이즈 등은 데이터에 따라 조정)
history = model.fit(X_train, y_train, epochs=50, batch_size=32, 
                    validation_split=0.2, verbose=1)

# %%
# 테스트 데이터 평가
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_acc:.4f}")

# %% [markdown]
# ## 4. 사용자 입력에 따른 예측 및 결과 시각화
#
# 아래 함수는 사용자 입력(딕셔너리 형태)을 받아 전처리한 후 모델 예측 결과(각 incidenttype별 확률)를 반환하고,  
# 이를 막대그래프로 시각화합니다.

# %%
def predict_incident(user_input):
    """
    사용자 입력 예시: 
    {
        "state_code": "WI",
        "month": 2,
        "max_temp": 23,
        "min_temp": -18,
        "precipitation": 5.0
    }
    """
    # 입력 딕셔너리를 DataFrame으로 변환
    df_input = pd.DataFrame([user_input])
    # 컬럼명 변환: user_input의 'state_code' -> 'state', 'max_temp' -> 'maxtemp', 'min_temp' -> 'mintemp'
    df_input = df_input.rename(columns={
        'state_code': 'state',
        'max_temp': 'maxtemp',
        'min_temp': 'mintemp'
    })
    
    # 상태(state)는 LabelEncoder를 사용하여 변환
    try:
        df_input['state'] = le_state.transform(df_input['state'])
    except Exception as e:
        print("입력 state가 학습 데이터에 없는 값입니다.", e)
        return None

    # 숫자형 피처 스케일링 (컬럼 순서는 학습 데이터와 동일해야 함)
    numeric_cols = ['month', 'maxtemp', 'mintemp', 'precipitation']
    df_input[numeric_cols] = scaler.transform(df_input[numeric_cols])
    
    # 모델 예측 (확률)
    pred_prob = model.predict(df_input)
    incident_probs = {incident: prob for incident, prob in zip(le_incident.classes_, pred_prob[0])}
    return incident_probs

def plot_probabilities(prob_dict):
    """ incident type 별 예측 확률을 막대그래프로 시각화 """
    if prob_dict is None:
        return
    labels = list(prob_dict.keys())
    probs = [v*100 for v in prob_dict.values()]  # 퍼센트 단위 변환
    
    plt.figure(figsize=(8,6))
    bars = plt.bar(labels, probs, color='skyblue')
    plt.xlabel('Incident Type')
    plt.ylabel('Probability (%)')
    plt.title('Predicted Incident Type Probabilities')
    plt.ylim(0, max(probs)*1.1)
    for bar, p in zip(bars, probs):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{p:.2f}%', ha='center', va='bottom')
    plt.show()

# %%
# 사용자 예시 입력
user_input = {
    "state_code": "MO",
    "month": 2,
    "max_temp": 90,
    "min_temp": 50,
    "precipitation": 30.0
}

# 예측 수행 및 결과 시각화
incident_probabilities = predict_incident(user_input)
print("예측 확률:")
for incident, prob in incident_probabilities.items():
    print(f"{incident}: {prob*100:.2f}%")

plot_probabilities(incident_probabilities)

# %%
# 사용자 예시 입력
user_input = {
    "state_code": "MN",
    "month": 3,
    "max_temp": 120,
    "min_temp": 80,
    "precipitation": 0.0
}

# 예측 수행 및 결과 시각화
incident_probabilities = predict_incident(user_input)
print("예측 확률:")
for incident, prob in incident_probabilities.items():
    print(f"{incident}: {prob*100:.2f}%")

plot_probabilities(incident_probabilities)
```
